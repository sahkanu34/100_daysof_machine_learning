# 12 Converters

```python
def rename(name):
    if name == "Machine Learning Engineer":
        return "MLE"
    else:
        return name 
```

```python
rename("Machine Learning Engineer")
```

```python
pd.read_csv('salaries.csv',converters={'job_title':rename}).head()
```

# 14 Handling Big data and break into chunks
